```python
# importing required libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing csv data

savings_df = pd.read_csv('savings_data.csv')
savings_df.head(10)
```

```python
# Checking the data
savings_df.info()
```

```python
# checking summary statistics
savings_df.describe()
```

```python
# Finding missing values per column.
savings_df.isnull().sum()
```

##### There are two missing values in the `Amount` column

```python
# Checking the missing values in the Amount column
savings_df[savings_df['Amount'].isnull()]
```

```python
# Filling the missing values with the median 
median_amt = savings_df['Amount'].median()
savings_df['Amount'] = savings_df['Amount'].fillna(median_amt)

# Checking missing values after fix
print("Missing values after fix: \n", savings_df.isnull().sum())
```

```python
# Checking for duplicate data

print("Duplicated rows found: ", savings_df.duplicated().sum())
```

```python
# Fixing data types

# Changing the month to datetime
savings_df['Month'] = pd.to_datetime(savings_df['Month'])
savings_df.info()
```

```python
# Saving to cleaned csv file
savings_df.to_csv('savings_data_clean.csv', index = False)
print('Data cleaning complete')
```
