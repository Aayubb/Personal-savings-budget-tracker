```python
import pandas as pd

monthly = pd.read_csv('csv_files/monthly_summary.csv', parse_dates=['Month'], index_col='Month')
category_breakdown = pd.read_csv('csv_files/category_breakdown.csv', index_col='Category').squeeze()

total_income = monthly['Total_income'].sum()
total_expenses = monthly['Total_expense'].sum()
total_saved = monthly['Total_savings'].sum()
avg_savings_rate = monthly['Savings_Rate(%)'].mean()
best_month = monthly['Savings_Rate(%)'].idxmax().strftime('%B %Y')
best_rate = monthly['Savings_Rate(%)'].max()
top_expense_cat = category_breakdown.idxmax()
year_end_balance = monthly['Running_balance'].iloc[-1]

print(total_income)
print(total_expenses)
print(total_saved)
print(avg_savings_rate)
print(best_month)
print(best_rate)
print(top_expense_cat)
print(year_end_balance)
```

## Executive Summary

This report analyzes 12 months of personal financial data for the year 2024.
The analysis covers income, expenditure, savings behavior, and cumulative
balance growth with the objective of demonstrating sound financial habits
and data analysis capability.

---

## Findings

**1. Consistent savings habit:**
The account holder saved an average of 50.55 % of monthly income
throughout the year, exceeding the recommended 20% savings benchmark in most months.

**2. Largest expense is rent:**
Rent accounted for the single largest expense category at Rs 144000
annually, which is typical for Dharan living costs. This is a fixed cost with
limited reduction potential.

**3. Salary increment impact:**
A salary increase in May 2024 (from Rs 45,000 to Rs 48,000) positively impacted
the savings rate in Q3, demonstrating income sensitivity to savings capacity.

**4. Variable spending patterns:**
Entertainment and transport costs showed the highest month-to-month variation.
partly driven by athletic training (marathon and triathlon)-related expenses.
These are discretionary and controllable.

---

## Recommendations

1. Reduce entertainment spending to below Rs 1,000/month to increase the savings rate by ~2%.
2. Build a 3-month emergency fund target of Rs 82350.
3. Consider a recurring deposit (RD) scheme at a local bank to earn interest on monthly savings.
