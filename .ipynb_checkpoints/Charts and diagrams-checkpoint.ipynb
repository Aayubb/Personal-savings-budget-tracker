```python
# Importing required libraries and csv files

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

monthly = pd.read_csv('csv_files/monthly_summary.csv', parse_dates=['Month'], index_col = 'Month')
category_breakdown = pd.read_csv('csv_files/category_breakdown.csv', index_col = 'Category').squeeze()

month_labels = monthly.index.strftime('%b')
```

```python
monthly.head()
```

```python
category_breakdown.head()
```

#### Build the bar chart and pie chart

##### Bar chart — expenses by category

```python
fig, ax = plt.subplots(figsize=(10, 5))

bars = ax.bar(category_breakdown.index, category_breakdown.values, color='steelblue', edgecolor='white')

# Add value labels on top of each bar
for bar in bars:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 200,
            f'Rs {bar.get_height():,.0f}', ha='center', va='bottom', fontsize=9)

ax.set_title('Annual Expenses by Category', fontsize=14, fontweight='bold')
ax.set_xlabel('Category')
ax.set_ylabel('Total Amount (NPR)')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'Rs {x:,.0f}'))
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig('charts/chart1_expenses_bar.png', dpi=150)
plt.show()
```

##### Pie chart — savings vs total spending

```python
total_savings = monthly['Total_savings'].sum()
total_expenses = monthly['Total_expense'].sum()

labels = ['Savings', 'Expenses']
values = [total_savings, total_expenses]
colors = ['#2ecc71', '#e74c3c']
explode = (0.05, 0)

fig, ax = plt.subplots(figsize=(7, 7))
wedges, texts, autotexts = ax.pie(
    values, labels=labels, colors=colors, explode=explode,
    autopct='%1.1f%%', startangle=90, textprops={'fontsize': 13}
)
ax.set_title('Savings vs Total Spending (Full Year)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('chart2_savings_pie.png', dpi=150)
plt.show()
```

#### Build the line chart and combine all charts into one figure

##### Line chart — running savings balance over months

```python
fig, ax = plt.subplots(figsize=(11, 5))

ax.plot(month_labels, monthly['Running_balance'], marker='o', linewidth=2.5,
        color='#2980b9', markersize=7, markerfacecolor='white', markeredgewidth=2)

# Annotate final balance
final_balance = monthly['Running_balance'].iloc[-1]
ax.annotate(f'Year-end: Rs {final_balance:,.0f}',
            xy=(month_labels[-1], final_balance),
            xytext=(-80, 15), textcoords='offset points',
            fontsize=10, color='#2980b9',
            arrowprops=dict(arrowstyle='->', color='#2980b9'))

ax.fill_between(range(len(month_labels)), monthly['Running_balance'].values, alpha=0.1, color='#2980b9')
ax.set_xticks(range(len(month_labels)))
ax.set_xticklabels(month_labels)
ax.set_title('Cumulative Savings Balance — 2024', fontsize=14, fontweight='bold')
ax.set_xlabel('Month')
ax.set_ylabel('Running Balance (NPR)')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'Rs {x:,.0f}'))
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('chart3_savings_line.png', dpi=150)
plt.show()
```

##### Combining all 3 charts into one professional figure

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Personal Financial Summary — 2024', fontsize=16, fontweight='bold', y=1.02)

# Chart 1 - Bar
axes[0].bar(category_breakdown.index, category_breakdown.values, color='steelblue', edgecolor='white')
axes[0].set_title('Expenses by Category')
axes[0].set_ylabel('NPR')
axes[0].tick_params(axis='x', rotation=30)

# Chart 2 - Pie
axes[1].pie([total_savings, total_expenses], labels=['Savings', 'Expenses'],
            colors=['#2ecc71', '#e74c3c'], autopct='%1.1f%%', startangle=90)
axes[1].set_title('Savings vs Spending')

# Chart 3 - Line
axes[2].plot(month_labels, monthly['Running_balance'], marker='o', color='#2980b9', linewidth=2)
axes[2].set_title('Running Savings Balance')
axes[2].set_ylabel('NPR')
axes[2].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.savefig('financial_summary_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
```
