```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Importing clean csv file for analysis
savings_df_clean = pd.read_csv('savings_data_clean.csv', parse_dates=['Month'])
savings_df_clean.head()
```

```python
# Checking for missing values/ duplicates
savings_df_clean.isnull().sum()
```

```python
savings_df_clean.duplicated().sum()
```

##### Calculating monthly totals and savings rate

```python
# Checking the type of money

savings_df_clean['Type'].unique()
```

```python
# Calculating monthly totals
# Seperating income, expense and savings.
income_df = savings_df_clean[savings_df_clean['Type'] == 'Income']
expense_df = savings_df_clean[savings_df_clean['Type'] == 'Expense']
savings_df = savings_df_clean[savings_df_clean['Type'] == 'Savings']

#Grouping them by month and sum
monthly_income = income_df.groupby('Month')['Amount'].sum().rename('Total_income')
monthly_expense = expense_df.groupby('Month')['Amount'].sum().rename('Total_expense')
monthly_savings = savings_df.groupby('Month')['Amount'].sum().rename('Total_savings')

# Mergint into one summary table
monthly = pd.concat([monthly_income, monthly_expense, monthly_savings], axis=1)
print(monthly)
```

```python
# Calculating Savings rate

# Savings rate = (Savings / Income) * 100
monthly['Savings_Rate(%)'] = (monthly['Total_savings'] / monthly['Total_income'] * 100).round(2)

print("\nMonthly Savings Rate:")
print(monthly[['Total_income', 'Total_savings', 'Savings_Rate(%)']])
```

```python
# Highest saving rate
monthly['Savings_Rate(%)'].max()
```
